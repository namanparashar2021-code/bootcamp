```python
# student=[{"name":"Naman","marks":90,"lab_assign":True,"city":"Agra","sec":"CB"},{"name":"Aman","marks":90,"lab_assign":True,"city":"Mathura","sec":"CB"},{"name":"Deepak","marks":92,"lab_assign":True,"city":"Mathura","sec":"CB"},{"name":"Akash","marks":95,"lab_assign":True,"city":"Agra","sec":"CB"}]

students = [
    {"name": "Naman", "marks": 92, "attendance": 95, "lab_assignment": True},
    {"name": "Aman", "marks": 95, "attendance": 100, "lab_assignment": True},
    {"name": "Deepak", "marks": 58, "attendance": 60, "lab_assignment": True},
    {"name": "Akash", "marks": 88, "attendance": 90, "lab_assignment": False},
    {"name": "Bhanu", "marks": 35, "attendance": 80, "lab_assignment": True}
]

for student in students:
    if student["lab_assignment"] == False:
        result = "Lab Assignment Not Submitted"
    elif student["attendance"] < 75:
        result = "Low Attendance"
    elif student["marks"] >= 70:
        result = "Distinction"
    elif student["marks"] >= 40:
        result = "Pass"
    else:
        result = "Fail"

    print(student["name"], "-", result)
```

```python
import pandas as pd

students_df = pd.DataFrame(students)
display(students_df.head())
```

```python
display(students_df.describe())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.barplot(x='name', y='marks', data=students_df, palette='viridis')
plt.title('Student Marks Distribution')
plt.xlabel('Student Name')
plt.ylabel('Marks')
plt.ylim(0, 100) # Assuming marks are out of 100
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
failed_assignments = students_df[students_df['lab_assignment'] == False]
print("Students who failed to submit their lab assignment:")
for name in failed_assignments['name']:
    print(name)
```

```python
submitted_assignments_df = students_df[students_df['lab_assignment'] == True]
average_attendance_submitted = submitted_assignments_df['attendance'].mean()
print(f"Average attendance for students who submitted their lab assignment: {average_attendance_submitted:.2f}%")
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='attendance', y='marks', data=students_df, hue='lab_assignment', s=100, alpha=0.8)
plt.title('Marks vs. Attendance')
plt.xlabel('Attendance (%)')
plt.ylabel('Marks')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
students_below_40_marks = students_df[students_df['marks'] < 40]
print("Students with marks below 40:")
if not students_below_40_marks.empty:
    for name in students_below_40_marks['name']:
        print(name)
else:
    print("No students found with marks below 40.")
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(students_df['marks'], bins=5, kde=True, color='skyblue')
plt.title('Distribution of Student Marks')
plt.xlabel('Marks')
plt.ylabel('Number of Students')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
correlation = students_df['marks'].corr(students_df['attendance'])
print(f"Correlation coefficient between Marks and Attendance: {correlation:.2f}")
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np

# Prepare the data
X = students_df['attendance'].values.reshape(-1, 1) # Independent variable (attendance)
y = students_df['marks'].values # Dependent variable (marks)

# Create and fit the linear regression model
model = LinearRegression()
model.fit(X, y)

# Print the coefficients
print(f"Intercept: {model.intercept_:.2f}")
print(f"Coefficient for Attendance: {model.coef_[0]:.2f}")
```

```python
r_squared = model.score(X, y)
print(f"R-squared value: {r_squared:.2f}")
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='attendance', y='marks', data=students_df, hue='lab_assignment', s=100, alpha=0.8)

# Plot the regression line
plt.plot(X, model.predict(X), color='red', linestyle='--', linewidth=2, label=f'Regression Line (R^2 = {r_squared:.2f})')

plt.title('Marks vs. Attendance with Regression Line')
plt.xlabel('Attendance (%)')
plt.ylabel('Marks')
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Re-defining students and students_df for self-contained execution
students = [
    {"name": "Naman", "marks": 92, "attendance": 95, "lab_assignment": True},
    {"name": "Aman", "marks": 95, "attendance": 100, "lab_assignment": True},
    {"name": "Deepak", "marks": 58, "attendance": 60, "lab_assignment": True},
    {"name": "Akash", "marks": 88, "attendance": 90, "lab_assignment": False},
    {"name": "Bhanu", "marks": 35, "attendance": 80, "lab_assignment": True}
]
students_df = pd.DataFrame(students)

plt.figure(figsize=(8, 6))
sns.boxplot(x='lab_assignment', y='marks', data=students_df, palette='pastel')
plt.title('Comparison of Marks Based on Lab Assignment Submission')
plt.xlabel('Lab Assignment Submitted (True/False)')
plt.ylabel('Marks')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
mean_marks_by_lab_assignment = students_df.groupby('lab_assignment')['marks'].mean().reset_index()
mean_marks_by_lab_assignment.rename(columns={'lab_assignment': 'Lab Assignment Submitted', 'marks': 'Mean Marks'}, inplace=True)
display(mean_marks_by_lab_assignment)
```

```python
from scipy import stats

# Extract marks for each group
marks_submitted = submitted_assignments_df['marks']
marks_missed = failed_assignments['marks']

# Perform independent samples t-test
t_statistic, p_value = stats.ttest_ind(marks_submitted, marks_missed, equal_var=False) # Assuming unequal variances

print(f"T-statistic: {t_statistic:.2f}")
print(f"P-value: {p_value:.3f}")

# Interpret the results
alpha = 0.05 # Significance level
if p_value < alpha:
    print("\nBased on the t-test, there is a statistically significant difference in marks between students who submitted lab assignments and those who missed them.")
else:
    print("\nBased on the t-test, there is no statistically significant difference in marks between students who submitted lab assignments and those who missed them.")
```
